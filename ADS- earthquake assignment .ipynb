```python
df.groupby(['Country', 'Rate units']).size()[:5]
```

```python
new_df=(df.groupby(['Country', 'Rate units']).size() .sort_values(ascending=False) .reset_index(name='count') )[:6]
new_df
```

```python
labels=new_df["Country"]+ " " + new_df["Rate units"].map(str)
sizes=new_df['count']
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.show()
```
